Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
#finding the highest score
highest_score=max(employees.values())
top_performers=[]
#finding employee who has highest score
for name,score in employees.items():
    if score==highest_score:
        top_performers.append(name)
print("Top Performers Eligible for Bonus:", ",".join(top_performers),"(Score:",highest_score,")")
```

Problem Statement 2: Search Query Keyword Analysis


```python
query="Buy mobile phone buy phone online"

#converting to lowercase
query=query.lower()

#remove punctuation
import string
query=query.translate(str.maketrans("","",string.punctuation))

#split words
words=query.split()

#count frequency
word_count={}
for word in words:
    word_count[word]=word_count.get(word,0)+1
    
#display words searched more than once
    result={}
    for word,count in word_count.items():
        if count>1:
            result[word]=count
print(result)
```

Problem Statement 3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Create empty list to store valid readings
valid_readings = []

# Loop through the list using index
for i in range(len(sensor_readings)):

# Check if the reading is even
if sensor_readings[i] % 2 == 0:

    # Store as hour_index, reading_value
     valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis


```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
#to store domain count
domain_count={}

#count domains
for email in emails:
    domain=email.split("@")[1]
    domain_count[domain]=domain_count.get(domain,0)+1
total_emails=len(emails)

#calculate and display percentage
for domain, count in domain_count.items():
    percentage=(count/total_emails)*100
    print(domain+":",str(int(percentage))+"%")
        
      
     
```

Problem Statement 5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

#calucalte avg  sales
total=0
for value in sales:
    total+=value
avg=total/len(sales)

#finding sales above average 30%
threshold=avg+500
for i in range(len(sales)):
    if sales[i]>=threshold:
        print("Day",i+1,":",sales[i])
```

Problem Statement 6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
user_count={}

#count occurrences
for user in user_ids:
    user_count[user]=user_count.get(user,0)+1

for user, count in user_count.items():
    if count>1:
        print(user," -> ",count,"times")
```
