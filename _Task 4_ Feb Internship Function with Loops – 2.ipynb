Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans=[199,299,399,599]

    while True:
        amount=int(input("Enter recharge amount:"))
        if amount<50:
            print("Minimum recharge amount is 50")
        elif amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid plan")
            print("Available plans are:",valid_plans)
            choice=input("Do you want to try again? (yes/no): ")
            if choice.lower()!="yes":
                print("Thank you")
                break
validate_recharge()
```

Problem 2: Inventory Reorder Alert System


```python
def check_stock(items):

    # Loop through each product in the dictionary
    for product in items:

        # Check if stock is less than 15
        if items[product]<15:
            print(product,"-Reoder Alert")   # Stock is low
        else:
            print(product,"-Stock OK")      # Stock is sufficient

# Dictionary containing product names and their stock quantities        
product={
    "Rice": 20,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 5
}
check_stock(product)
```

Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0

    # Calculate total marks using loop
    for mark in marks:
        total += mark
    # Find average
    average = total / len(marks)

    print("Average Marks:", average)
     # Check pass or fail
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Sample marks list
marks = [60, 55, 70, 45]
result = process_result(marks)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option


```python
def fare_calculator(distance,peak):

    # Base fare ₹50 + ₹12 per km
    fare=50+(distance*12)

     # Add 25% extra during peak hour
    if peak=="yes":
        fare=fare+(fare*0.25)

    return fare

while True:
    km=float(input("Enter distance in km:"))
    peak_time=input("Is it peak hour?(yes/no):")

    total=fare_calculator(km,peak_time.lower())
    print("Total Fare: Rs. ",total)

    # Ask user if they want to calculate again
    again=input("Do you want to calculate again?(yes/no):")
    if again.lower()!="yes":
        print("Done")
        break
```

Problem 5: Employee Attendance Eligibility Checker


```python
def attendance_check(attendance_list):
    present=0

    # Count number of present days
    for day in attendance_list:
        if day=="P":
            present=present+1

    # Calculate attendance percentage        
    percentage=(present/len(attendance_list))*100
    print("Attendance percentage:",percentage)

    # Check eligibility
    if percentage>=75:
        print("Eligible")
    else:
        print("Not Eligible")
# Sample attendance list
attendance=["P", "A", "P", "P", "P", "P", "P"]
attendance_check(attendance)              
```

Problem 6: Password Strength Checker


```python
def check_password(password):

    # Check minimum length
    if len(password) < 8:
        print("Weak Password")
        return

    has_digit = False
    has_special = False

    # Check for digit and special character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    # Final strength check
    if has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")

pwd = input("Enter password: ")
check_password(pwd)
```
