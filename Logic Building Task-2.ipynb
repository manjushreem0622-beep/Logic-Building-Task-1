Problem 1: Unique Words in a Sentence


```python
#Input sentence
sen="python is easy and python is powerful"
#split sentence into words
words=sen.split()
unique_words=set(words)
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
#employee dictionary
employees={
     "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
#assuming first employee has high salary
highest_salary=0
highest_employee=""

#comparing salary
for name in employees:
    if employees[name]>highest_salary:
        highest_salary=employees[name]
        highest_employee=name
        
print("Highest salary:",highest_employee," ",highest_salary)
```

Problem 3: Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 10, 66]

#initializing the values
max_value=numbers[0]
min_value=numbers[0]

#comparing each other
for n in numbers:
    if n>max_value:
        max_value=n
    if n<min_value:
        min_value=n
        
print("List:",numbers)
print("Max:",max_value)
print("Min:",min_value)
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

#counter
count =0

#check for each price
for price in prices:
    if price>1000:
        count+=1

print("Products above 1000:",count)
```

Problem 5: Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]

present_count=0

#count p
for status in attendance:
    if status=="P":
        present_count+=1
total_days=len(attendance)

#calculate percentage
percentage=(present_count/total_days)*100

print("Attendance Percentage:",percentage)
```

Problem 7: Count Character Frequency


```python
text="python"

#empty dictionary
frequency={}

#cunt each character
for ch in text:
    if ch in frequency:
        frequency[ch]+=1
    else:
        frequency[ch]=1

print(frequency)
```
