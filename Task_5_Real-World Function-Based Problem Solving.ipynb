Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes=0

    # Calculate total likes using loop
    for like in likes_list:
        total_likes+=like

    print("Total Likes:",total_likes)

        # Check if post is viral
    if total_likes>=1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

# Example data
likes=[200,300,250,300]
analyze_engagement(likes)
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_check(rainfall_list,required_level):
    total=0

    # Calculate total rainfall
    for rain in rainfall_list:
        total+=rain

    avg=total/len(rainfall_list)
    print("Average Rainfall:",avg)

        # Compare with required level
    if avg>=required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall_data=[60,70,80,78]
rainfall_check(rainfall_data,65)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def check_duplicates(usernames):

    # Convert list to set
    if len(usernames)!=len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users= ["manju","ramya","manju","shree"]
check_duplicates(users)
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def attendance_check(attendance_list):
    present=0

     # Count present days
    for day in attendance_list:
        if day=="P":
            present+=1

    #attendance percentage
    percentage=(present/len(attendance_list))*100
    print("Attendance Percentage:", percentage)

    #checking for attendance eligibility
    if percentage>=75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance=["P","P","A","P","A","P","P","P","P","P"]
attendance_check(attendance)
```

Problem 10: Smart Electricity Bill Analyzer


```python
def calculate_bill(units):
    bill=0

    # Slab calculation
    if units<=100:
        bill=units*3
    elif units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)

    # Classify usage    
    if bill<500:
        status= "Low Usage"
    elif bill<=1500:
        status= "Moderate Usage"
    else:
        status= "High Usage"
    return bill,status

total_bill,usage_status=calculate_bill(700)
print("Total Bill:",total_bill)
print("Usage Status:", usage_status)
```
