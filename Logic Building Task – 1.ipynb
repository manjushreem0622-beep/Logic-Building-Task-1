```python
marks=[45, 78, 90, 33, 60]
pass_students=0
fail_students=0
for i in marks:
    if i>=50:
        pass_students+=1
    else:
        fail_students+=1
print("No. of students passed:",pass_students)
print("No. of students failed:",fail_students)

```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned=[]
for n in names:
    n=n.strip()
    n=n.lower()
    cleaned.append(n)
print("Cleaned names:",cleaned)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for m in messages:
    m_len=len(m)
    print("Message:",m)
    print("Length:",m_len)
    if m_len>10:
        print("--Flag:Message is longer than 10 characters")

```
